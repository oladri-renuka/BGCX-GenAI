```python
import pandas as pd
```

```python
df=pd.read_csv(r'company.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df.head(9)
```

```python
df.fillna(method='bfill', inplace=True)
```

```python
df.head(9)
```

```python
growth_aggregated_df = df.groupby(['Company', 'Fiscal Year']).agg(
    {
        'Revenue Growth (%)': 'mean',
        'Net Income Growth (%)': 'mean',
        'Assets Growth (%)': 'mean',
        'Liabilities Growth (%)': 'mean',
        'Cash Flow Growth (%)': 'mean'
    }
).reset_index()

# Rename columns for clarity
growth_aggregated_df.columns = [
    'Company', 'Fiscal Year', 
    'Avg Revenue Growth (%)', 
    'Avg Net Income Growth (%)', 
    'Avg Assets Growth (%)', 
    'Avg Liabilities Growth (%)', 
    'Avg Cash Flow Growth (%)'
]
```

### **Summary of Findings**

The analysis of year-over-year growth across key financial metrics for three major companies—Apple, Microsoft, and Tesla—reveals distinct patterns in their financial performance from 2021 to 2023.

1. **Apple**:
   - **Revenue and Net Income**: Apple showed moderate growth in revenue and net income in 2021 and 2022, with an average growth rate of approximately 7.8% and 5.4%, respectively. However, in 2023, both revenue and net income declined by around 2.8%. This suggests a slowdown in Apple's financial performance, possibly due to market saturation, economic conditions, or competitive pressures.
  
  - **Assets and Liabilities**: The company's assets and liabilities remained relatively stable, with slight growth in 2021 and 2022, followed by a small decline in 2023. This could indicate careful management of its financial structure, though the decline in assets and liabilities in 2023 could be a sign of decreased investment or asset disposal.
  
  - **Cash Flow**: Cash flow from operating activities saw significant growth in 2021 and 2022 but dropped sharply by 9.5% in 2023, reflecting potential challenges in maintaining operating efficiency or changes in working capital requirements.

2. **Microsoft**:
   - **Revenue**: Microsoft experienced strong revenue growth in 2021 and 2022, with an average increase of about 18%. However, the growth rate dropped to 6.9% in 2023, indicating a potential deceleration in sales or other income-generating activities.
   - **Net Income**: The net income growth mirrored the revenue trend, with a significant increase in 2021 and 2022, followed by a minor decline in 2023. This suggests that Microsoft's profitability was slightly impacted in 2023 despite continued revenue growth.
   - **Assets and Liabilities**: Microsoft’s assets grew steadily across the years, with the highest growth in 2023 (12.9%), possibly due to acquisitions or new investments. Liabilities also increased but at a slower rate, indicating a strengthening balance sheet.
   - **Cash Flow**: Cash flow from operating activities remained strong in 2021 and 2022, but there was a slight negative growth in 2023 (-1.6%), possibly indicating tighter cash flows despite increased assets.

3. **Tesla**:
   - **Revenue**: Tesla exhibited remarkable revenue growth in 2021 and 2022, with a staggering 51.4% increase each year. This reflects Tesla's rapid expansion and increased market penetration during this period. However, the growth rate moderated to 18.8% in 2023, which, while still strong, shows signs of maturing growth.
   - **Net Income**: Tesla’s net income saw an extraordinary growth of over 127% in 2021 and 2022, reflecting improved profitability. This dropped to 19.3% in 2023, which is still impressive but much lower than the previous years.
   - **Assets and Liabilities**: Tesla’s assets grew significantly, especially in 2021 and 2022, indicating heavy investment in production capacity and other assets. Liabilities also grew but were well-managed relative to the increase in assets.
   - **Cash Flow**: Tesla's cash flow from operating activities grew strongly in 2021 and 2022, consistent with its revenue and net income growth. However, in 2023, it decreased by almost 10%, possibly due to increased capital expenditures or changes in working capital requirements.

### **Conclusion**
- **Growth Slowdown**: Both Apple and Microsoft experienced a slowdown in revenue and net income growth in 2023, whereas Tesla, though still growing, also showed signs of deceleration. This trend could indicate broader economic factors affecting these companies.
- **Investment and Assets**: Microsoft and Tesla continued to grow their assets, with Tesla showing particularly strong asset growth, indicating continued investment in infrastructure and capacity.
- **Cash Flow Challenges**: All three companies saw declines in cash flow growth in 2023, suggesting potential challenges in managing operating cash flows despite overall growth in previous years.
